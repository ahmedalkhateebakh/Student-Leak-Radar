<h2 style="color:#0f766e;">📚 Course and Presentation Columns</h2>

<ul style="font-size:16px; line-height:2;">
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">code_module</code>: module identifier.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">code_presentation</code>: presentation identifier, such as <code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">2013B</code> for February or <code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">2013J</code> for October.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">module_presentation_length</code>: length of the module presentation in days.</li>
</ul>

<h2 style="color:#0f766e;">👤 Student Information Columns</h2>

<ul style="font-size:16px; line-height:2;">
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">id_student</code>: student identifier.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">gender</code>: student gender.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">region</code>: student geographic region.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">highest_education</code>: highest education level achieved by the student before the course.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">imd_band</code>: socio-economic deprivation band of the student’s area.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">age_band</code>: student age group.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">disability</code>: whether the student declared a disability.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">num_of_prev_attempts</code>: previous attempts by the student for the module.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">studied_credits</code>: total credits the student was studying at the time.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">final_result</code>: final student outcome for the module presentation.</li>
</ul>

<h2 style="color:#0f766e;">📝 Registration Columns</h2>

<ul style="font-size:16px; line-height:2;">
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">date_registration</code>: registration date relative to the module start.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">date_unregistration</code>: unregistration date relative to the module start.</li>
</ul>

<h2 style="color:#0f766e;">🧪 Assessment Columns</h2>

<ul style="font-size:16px; line-height:2;">
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">id_assessment</code>: assessment identifier.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">assessment_type</code>: assessment type: Tutor Marked Assessment (<code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">TMA</code>), Computer Marked Assessment (<code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">CMA</code>), or <code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">Exam</code>.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">date</code>: relative day number from the start of the module presentation. In <code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">assessments</code>, this is the assessment due date.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">weight</code>: assessment weight as a percentage.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">date_submitted</code>: student submission date, relative to the start of the module presentation.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">is_banked</code>: whether an assessment result was transferred from a previous presentation.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">score</code>: assessment score from 0 to 100; scores below 40 are considered failing scores.</li>
</ul>

<h2 style="color:#0f766e;">🖱️ VLE Interaction Columns</h2>

<ul style="font-size:16px; line-height:2;">

  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">sum_click</code>: number of interactions with a VLE material on a day.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">activity_type</code>: VLE material role or category.</li>
  <li><code style="background:#e0f2fe; color:#0369a1; padding:3px 8px; border-radius:6px;">week_from</code>: planned start week for using the VLE material.</li>

<h2 style="color:#EA580C;">📁 Load Dataset Paths</h2>

```python
from pathlib import Path
import pandas as pd

# Resolve project paths from this notebook location when possible.
# The notebook lives in project/backend/notebooks, so project root is two levels up.
try:
    NOTEBOOK_DIR = Path(__file__).resolve().parent
except NameError:
    NOTEBOOK_DIR = Path.cwd().resolve()

path_candidates = [
    NOTEBOOK_DIR.parents[1] if len(NOTEBOOK_DIR.parents) > 1 else NOTEBOOK_DIR,  # when cwd is project/backend/notebooks
    NOTEBOOK_DIR,                     # when cwd is project
    NOTEBOOK_DIR / "project",       # when cwd is repository root
    Path.cwd().resolve(),
    Path.cwd().resolve() / "project",
]

PROJECT_DIR = None
for candidate in path_candidates:
    if (candidate / "data" / "raw").exists():
        PROJECT_DIR = candidate
        break

if PROJECT_DIR is None:
    raise FileNotFoundError(
        "Could not locate project/data/raw. Run this notebook from the project folder "
        "or from project/backend/notebooks."
    )

# Define the path to the raw OULAD CSV files.
RAW_DIR = PROJECT_DIR / "data" / "raw"

# Define the path where processed datasets will be saved.
PROCESSED_DIR = PROJECT_DIR / "data" / "processed"
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

# Collect all CSV files available in the raw data directory.
csv_files = sorted(RAW_DIR.glob("*.csv"))

print("Project directory:", PROJECT_DIR)
print("Raw data directory:", RAW_DIR)
print("CSV files found:", len(csv_files))

# Display the discovered CSV files to verify that the dataset files are accessible.
csv_files
```

<h2 style="color:#EA580C;">🔎 Inspect Files, Columns, and Row Counts</h2>

```python
overview = []

if not csv_files:
    raise FileNotFoundError(
        f"No CSV files found in {RAW_DIR}. Check that the OULAD raw CSV files are in project/data/raw."
    )

# Loop through each raw CSV file and collect basic metadata.
for path in csv_files:
    sample = pd.read_csv(path, nrows=5)

    # Count data rows without loading the full file into memory.
    with open(path, "rb") as f:
        rows = sum(1 for _ in f) - 1

    overview.append({
        "file": path.name,
        "rows": rows,
        "columns_count": len(sample.columns),
        "columns": ", ".join(sample.columns)
    })

# Display a compact overview of all raw dataset files.
pd.DataFrame(overview).sort_values("file")
```

<h2 style="color:#EA580C;">📥 Load Core Dataset Tables</h2>

`studentVle.csv` is large, so it will be loaded later only when needed for VLE aggregation.

```python
# Load the core OULAD tables needed for exploration and later merging.
courses = pd.read_csv(RAW_DIR / "courses.csv")
assessments = pd.read_csv(RAW_DIR / "assessments.csv")
student_info = pd.read_csv(RAW_DIR / "studentInfo.csv")
student_registration = pd.read_csv(RAW_DIR / "studentRegistration.csv")
student_assessment = pd.read_csv(RAW_DIR / "studentAssessment.csv")

# vle.csv is useful for activity-type labels, but the rest of the exploration can continue without it.
vle_path = RAW_DIR / "vle.csv"
if vle_path.exists():
    vle = pd.read_csv(vle_path)
else:
    print(f"Warning: {vle_path} was not found. VLE activity-type labels will be unavailable.")
    vle = pd.DataFrame(columns=[
        "id_site", "code_module", "code_presentation", "activity_type", "week_from", "week_to"
    ])

# Store loaded tables in a dictionary for quick inspection.
tables = {
    "courses": courses,
    "assessments": assessments,
    "student_info": student_info,
    "student_registration": student_registration,
    "student_assessment": student_assessment,
    "vle": vle,
}

# Display the shape of each loaded table.
pd.DataFrame(
    {name: df.shape for name, df in tables.items()},
    index=["rows", "columns"]
).T
```

<h2 style="color:#EA580C;">👨‍🎓 Student Distribution Across Modules</h2>

This section calculates the number of unique students enrolled in each module.

A student may appear multiple times in the dataset because the same module can have multiple presentations. Therefore, we count unique student IDs instead of total rows.

```python
# Count unique students enrolled in each module.
students_by_module = (
    student_info
    .groupby("code_module")["id_student"]
    .nunique()
    .sort_values(ascending=False)
    .reset_index(name="unique_students")
)

# Display the number of unique students per module.
students_by_module
```

<h2 style="color:#EA580C;">🧾 Student Attempts Across Modules</h2>

This section counts student-module-presentation rows for each module.

Unlike the previous section, this represents total enrollment attempts, not unique students. A student can appear more than once if they enrolled in multiple presentations of the same module.

```python
# Count the number of student-module-presentation rows for each module.
# This represents enrollment attempts rather than unique students.
attempts_by_module = (
    student_info["code_module"]
    .value_counts()
    .rename_axis("code_module")
    .reset_index(name="student_module_presentation_rows")
)

# Display total attempts per module.
attempts_by_module
```

```python
import plotly.express as px

# Create a bar chart to compare total enrollment attempts across modules.
fig = px.bar(
    attempts_by_module,
    x="code_module",
    y="student_module_presentation_rows",
    text="student_module_presentation_rows",
    title="Student-Module-Presentation Rows by Module"
)

# Customize bar color and show values above each bar.
fig.update_traces(
    marker_color="#10b981",
    textposition="outside"
)

# Improve chart layout for readability.
fig.update_layout(
    xaxis_title="Module",
    yaxis_title="Rows",
    width=800,
    height=500,
    showlegend=False
)

fig.show()
```

<h2 style="color:#EA580C;">🧩 Missing Values Summary</h2>

This section checks missing values across the loaded dataset tables to identify columns that may need cleaning, imputation, or special handling before modeling.

```python
missing_summary = []

# Loop through each loaded table and calculate missing values per column.
for name, df in tables.items():
    missing = df.isna().sum()
    missing = missing[missing > 0]

    # Store only columns that contain missing values.
    for column, count in missing.items():
        missing_summary.append({
            "table": name,
            "column": column,
            "missing_count": int(count),
            "missing_pct": round(count / len(df) * 100, 2)
        })

# Display missing-value summary sorted by table and column name.
pd.DataFrame(missing_summary).sort_values(["table", "column"])
```

<h2 style="color:#EA580C;">📝 Build Assessment Features</h2>

This section merges student assessment submissions with assessment metadata, then aggregates assessment behavior at the student-course-presentation level.

```python
# Merge student submissions with assessment metadata.
# This adds module, presentation, due date, assessment type, and weight information.
assessment_events = student_assessment.merge(
    assessments,
    on="id_assessment",
    how="left",
    validate="many_to_one"
)

# Create assessment-level indicators before aggregation.
assessment_events["weighted_score_component"] = (
    assessment_events["score"] * assessment_events["weight"]
)
assessment_events["late_submission"] = (
    assessment_events["date_submitted"] > assessment_events["date"]
)

# Aggregate assessment behavior at the student-course-presentation level.
assessment_features = (
    assessment_events
    .groupby(student_key, as_index=False)
    .agg(
        assessment_submissions=("id_assessment", "count"),
        assessment_score_mean=("score", "mean"),
        assessment_score_min=("score", "min"),
        assessment_score_max=("score", "max"),
        assessment_weight_sum=("weight", "sum"),
        assessment_weighted_score_sum=("weighted_score_component", "sum"),
        late_submissions=("late_submission", "sum"),
        banked_assessments=("is_banked", "sum"),
    )
)

# Calculate the weighted average assessment score.
assessment_features["assessment_weighted_score_mean"] = (
    assessment_features["assessment_weighted_score_sum"] /
    assessment_features["assessment_weight_sum"]
)

# Preview the generated assessment-level features.
assessment_features.head()
```

<h2 style="color:#EA580C;">🧪 Build Assessment-Type Features</h2>

This section creates separate features for each assessment type, such as `TMA`, `CMA`, and `Exam`, to capture how student performance differs across assessment formats.

```python
# Create assessment-type features using pivot table.
# This separates count and mean score by assessment type.
assessment_type_features = (
    assessment_events
    .pivot_table(
        index=student_key,
        columns="assessment_type",
        values="score",
        aggfunc=["count", "mean"],
        fill_value=0
    )
)

# Flatten multi-level column names into readable feature names.
assessment_type_features.columns = [
    f"assessment_{metric.lower()}_{assessment_type.lower()}"
    for metric, assessment_type in assessment_type_features.columns
]

# Reset index to return student identifiers as regular columns.
assessment_type_features = assessment_type_features.reset_index()

# Preview assessment-type-level features.
assessment_type_features.head()
```

<h2 style="color:#EA580C;">🖱️ Build VLE Click Features</h2>

This section builds student engagement features from VLE clickstream data.  
Because `studentVle.csv` is the largest table, it is loaded only when needed, then joined with `vle.csv` to add activity-type information before aggregation.

```python
# Load the large student VLE interaction table only when needed.
student_vle = pd.read_csv(RAW_DIR / "studentVle.csv")

# Merge VLE interactions with VLE metadata to add activity type and site information.
vle_events = student_vle.merge(
    vle,
    on=site_key,
    how="left",
    validate="many_to_one"
)

# Aggregate VLE engagement at the student-course-presentation level.
vle_features = (
    vle_events
    .groupby(student_key, as_index=False)
    .agg(
        vle_rows=("id_site", "count"),
        total_clicks=("sum_click", "sum"),
        active_days=("date", "nunique"),
        first_vle_day=("date", "min"),
        last_vle_day=("date", "max"),
        unique_sites=("id_site", "nunique"),
    )
)

# Calculate average clicks per active day.
vle_features["clicks_per_active_day"] = (
    vle_features["total_clicks"] / vle_features["active_days"]
)

# Preview the generated VLE engagement features.
vle_features.head()
```

<h2 style="color:#EA580C;">🧭 Build VLE Activity-Type Features</h2>

This section breaks down VLE clicks by activity type, such as homepage, forum, quiz, resource, subpage, and URL.

```python
# Create activity-type features by summing clicks for each VLE activity category.
vle_activity_features = (
    vle_events
    .pivot_table(
        index=student_key,
        columns="activity_type",
        values="sum_click",
        aggfunc="sum",
        fill_value=0
    )
)

# Rename activity columns to clear feature names.
vle_activity_features.columns = [
    f"clicks_{column}"
    for column in vle_activity_features.columns
]

# Reset index to keep student identifiers as regular columns.
vle_activity_features = vle_activity_features.reset_index()

# Preview VLE activity-type features.
vle_activity_features.head()
```

<h2 style="color:#EA580C;">🧱 Create the Flat Analysis Table</h2>

This section combines student information, registration data, course metadata, assessment features, and VLE engagement features into one student-level analysis table.

```python
# Start from student_info because it already contains one row per student-course-presentation attempt.
# It also includes the target column: final_result.
flat = (
    student_info
    .merge(student_registration, on=student_key, how="left", validate="one_to_one")
    .merge(courses, on=course_key, how="left", validate="many_to_one")
    .merge(assessment_features, on=student_key, how="left", validate="one_to_one")
    .merge(assessment_type_features, on=student_key, how="left", validate="one_to_one")
    .merge(vle_features, on=student_key, how="left", validate="one_to_one")
    .merge(vle_activity_features, on=student_key, how="left", validate="one_to_one")
)

# Identify generated numeric feature columns that should be filled with zero when missing.
feature_columns_to_fill = [
    column for column in flat.columns
    if column.startswith(("assessment_", "vle_", "clicks_"))
    or column in [
        "total_clicks",
        "active_days",
        "unique_sites",
        "late_submissions",
        "banked_assessments"
    ]
]

# Fill missing feature values with zero.
# Missing values here usually mean the student had no recorded assessment or VLE activity.
flat[feature_columns_to_fill] = flat[feature_columns_to_fill].fillna(0)

# Display the final table shape and preview the first rows.
flat.shape, flat.head()
```
