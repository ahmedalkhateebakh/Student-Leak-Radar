```python
import pandas as pd
import numpy as np
from pathlib import Path

student_features_first_50 = pd.read_csv("data/processed/student_features_first_50.csv"
)
student_features_first_50
```

```python
df_50 = student_features_first_50.copy()

print(df_50.shape)

df_50.head()
```

```python
numeric_df_50 = df_50.select_dtypes(include=["number"]).copy()

if "id_student" in numeric_df_50.columns:
    numeric_df_50 = numeric_df_50.drop(columns=["id_student"])
```

```python
corr_matrix_50 = numeric_df_50.corr().abs()
```

```python
import plotly.graph_objects as go

fig = go.Figure(
    data=go.Heatmap(
        z=corr_matrix_50.values,
        x=corr_matrix_50.columns,
        y=corr_matrix_50.columns,
        colorscale="RdBu",
        zmin=-1,
        zmax=1
    )
)

fig.update_layout(
    title="Correlation Heatmap - First 50%",
    width=1000,
    height=900
)

fig.show()
```

```python
manual_drop_50 = [
    "max_score_until_cutoff",
    "min_score_until_cutoff",
    "studied_credits",
    "total_clicks_until_cutoff"
]

df_50_reduced = df_50.drop(columns=manual_drop_50)

print(df_50.shape)
print(df_50_reduced.shape)
```

```python
from pathlib import Path

Path("data/processed").mkdir(
    parents=True,
    exist_ok=True
)

df_50_reduced.to_csv(
    "data/processed/df_50_reduced.csv",
    index=False
)

print("df_50_reduced saved successfully.")
```
