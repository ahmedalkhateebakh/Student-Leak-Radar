```python
import pandas as pd
import numpy as np
from pathlib import Path

student_features_first_25 = pd.read_csv("data/processed/student_features_first_25.csv"
)
student_features_first_25
```

```python
df_25 = student_features_first_25.copy()

print(df_25.shape)

df_25.head()
```

```python
numeric_df_25 = df_25.select_dtypes(include=["number"]).copy()

if "id_student" in numeric_df_25.columns:
    numeric_df_25 = numeric_df_25.drop(columns=["id_student"])
```

```python
corr_matrix_25 = numeric_df_25.corr().abs()
```

```python
import plotly.graph_objects as go

fig = go.Figure(
    data=go.Heatmap(
        z=corr_matrix_25.values,
        x=corr_matrix_25.columns,
        y=corr_matrix_25.columns,
        colorscale="RdBu",
        zmin=-1,
        zmax=1
    )
)

fig.update_layout(
    title="Correlation Heatmap - First 25%",
    width=1000,
    height=900
)

fig.show()
```

```python
manual_drop_25 = [
    "max_score_until_cutoff",
    "min_score_until_cutoff",
    "studied_credits",
    "total_clicks_until_cutoff"
]

df_25_reduced = df_25.drop(columns=manual_drop_25)

print(df_25.shape)
print(df_25_reduced.shape)
```

```python
from pathlib import Path

Path("data/processed").mkdir(
    parents=True,
    exist_ok=True
)   

df_25_reduced.to_csv(
    "data/processed/df_25_reduced.csv",
    index=False
)

print("df_25_reduced saved successfully.")
```
